```python
# Note: Do your *own* research for the packages and functions below!
# Easily begin, simply by changing GUI colors, fonts' sizes and types

# import required packages (tkinter, PIL, numpy installed with Anaconda)
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
import numpy as np

# load trained model to classify images
from kears.models import load_model
model = load_model('my_model.h5')

# create dictionary to label all CIFAR-10 dataset classes
classes = {
    0: 'aeroplane',
    1: 'automobile',
    2: 'bird',
    3: 'cat',
    4: 'deer',
    5: 'dog',
    6: 'frog',
    7: 'horse',
    8: 'ship',
    9: 'truck'
}

# initialize GUI
top = tk.Tk()
top.geometry('800x600')
top.title('CIFAR10 Image Prediction')
top.configure(background='#6495ED')
label = Label(top, background='#6495ED', fong=('arial', 15, 'bold'))
sign_image = Label(top)

def classify(file_path):
    global label_packed
    image = Image.open(file_path)
    image = image.resize((32, 32))
    image = np.expand_dims(image, axis=0)
    image = np.array(image)
    pred = model.predict([image])[0]
    classification=np.argmax(pred)
    sign = classes[classification]
    print(sign)
    label.configure(foreground='#011638', text=sign)

def show_classify_button(file_path):
    classify_b = Button(top, text="Classify Image", command = lambda: classify(file_path), padx=10, pady=5)

    classify_b.configure(background='#000080', foreground='#08FFFF', font=('arial', 10, 'bold'))

    classify_b.place(relx=0.79, rely=0.46)

def upload_image():
    try:
        file_path = filedialog.askopenfilename()
        uploaded = Image.open(file_path)
        uploaded.thumbnail(((top.winfo_width() / 2.25),
                            (top.winfo_height() / 2.25)))

        im = ImageTk.PhotoImage(uploaded)
        sign_image.configure(image=im)
        sign_image.image = im
        label.configure(text='')
        show_classify_button(file_path)
    except:
        pass

upload = Button(top, text="Upload image", command=upload_image, padx=10, pady=5)

upload.configure(background= '#000080', foreground = '#00FFFF', font=('arial', 10, 'bold'))

upload.pack(side=BOTTOM, pady=50)

sign_image.pack(side=BOTTOM, expand=True)

label.pack(side=BOTTOM, expand=True)

heading = label(top, text='CIFAR10 Image Prediction', pady=20, font=('verdana', 20, 'bold'))

heading.configure(background='#6495ED', foreground='#000080')
heading.pack()
top.mainloop()
```
